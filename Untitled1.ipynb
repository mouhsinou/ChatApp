```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from skimage.measure import regionprops, label
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from tensorflow.keras.datasets import mnist
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
from sklearn.model_selection import GridSearchCV
from skimage import transform
import os
from PIL import Image
from scipy.ndimage import binary_dilation, label
from skimage.morphology import rectangle
import csv  # Ajout de l'importation manquante pour CSV
```

ETAPE DE PRETRAITEMENT

```python
# Fonction pour binariser l'image
def binarize_image(img, threshold):
    return img.point(lambda p: p > threshold and 255)

# Fonction pour combiner deux images binarisées
def combine_binarizations(binary_image1, binary_image2):
    binary_array1 = np.array(binary_image1)
    binary_array2 = np.array(binary_image2)
    combined_binary_array = np.logical_or(binary_array1, binary_array2)
    combined_binary_image = Image.fromarray(np.uint8(combined_binary_array) * 255)
    final_binary_image = Image.fromarray(np.invert(np.array(combined_binary_image)))
    return final_binary_image

# Fonction pour extraire les cavités d'une région donnée
def extract_cavities(labeled_image, region_label, output_dir):

    """
    Extrait les cavités d'une région étiquetée dans l'image et les enregistre.

    Paramètres:
    labeled_image (ndarray): L'image étiquetée.
    region_label (int): Le label de la région à extraire.
    output_dir (str): Le répertoire où enregistrer les cavités.

    Retourne:
    list: Liste des caractéristiques normalisées [Est, Ouest, Sud, Nord, Centre].
    """
    bin_region = (labeled_image == region_label)
    x, y = np.where(bin_region)

    max_x, max_y = np.max(x), np.max(y)
    min_x, min_y = np.min(x), np.min(y)

    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    # Cavité EST
    FS = rectangle(1, (max_y - min_y) // 2)
    cavite_est = binary_dilation(bin_region, structure=FS)
    a, b = np.where(cavite_est)
    ax = axes[0, 2]
    ax.imshow(cavite_est, cmap='gray')
    ax.set_title('Cavité EST')
    ax.set_xlim([min(b), (min(b) + max(b)) // 2])
    ax.set_ylim([min(a), max(a)])

    # Cavité OUEST
    FS = rectangle(1, (max_y - min_y) // 2)
    cavite_ouest = binary_dilation(bin_region, structure=FS)
    a, b = np.where(cavite_ouest)
    ax = axes[0, 1]
    ax.imshow(cavite_ouest, cmap='gray')
    ax.set_title('Cavité OUEST')
    ax.set_xlim([(min(b) + max(b)) // 2, max(b)])
    ax.set_ylim([min(a), max(a)])

    # Cavité SUD
    FS = rectangle((max_x - min_x) // 2, 1)
    cavite_sud = binary_dilation(bin_region, structure=FS)
    a, b = np.where(cavite_sud)
    ax = axes[1, 0]
    ax.imshow(cavite_sud, cmap='gray')
    ax.set_title('Cavité SUD')
    ax.set_xlim([min(b), max(b)])
    ax.set_ylim([min(a), (min(a) + max(a)) // 2])

    # Cavité NORD
    FS = rectangle((max_x - min_x) // 2, 1)
    cavite_nord = binary_dilation(bin_region, structure=FS)
    a, b = np.where(cavite_nord)
    ax = axes[0, 0]
    ax.imshow(cavite_nord, cmap='gray')
    ax.set_title('Cavité NORD')
    ax.set_xlim([min(b), max(b)])
    ax.set_ylim([(min(a) + max(a)) // 2, max(a)])

    plt.tight_layout()
    plt.show()

    # Sauvegarder les cavités dans un dossier
    cavite_est_img = Image.fromarray((cavite_est * 255).astype(np.uint8))
    cavite_ouest_img = Image.fromarray((cavite_ouest * 255).astype(np.uint8))
    cavite_sud_img = Image.fromarray((cavite_sud * 255).astype(np.uint8))
    cavite_nord_img = Image.fromarray((cavite_nord * 255).astype(np.uint8))

    # Sauvegarde des cavités dans le dossier
    cavite_est_img.save(os.path.join(output_dir, f'region_{region_label}_est.png'))
    cavite_ouest_img.save(os.path.join(output_dir, f'region_{region_label}_ouest.png'))
    cavite_sud_img.save(os.path.join(output_dir, f'region_{region_label}_sud.png'))
    cavite_nord_img.save(os.path.join(output_dir, f'region_{region_label}_nord.png'))

    # Enfin obtenir les vecteurs d'attributs de chaque chiffre qui peut être normalisé ( [Est, Ouest, Sud, Nord, Centre])
    numerisation_cavite_est = np.array(cavite_est * 255)
    numerisation_cavite_ouest = np.array(cavite_ouest * 255)
    numerisation_cavite_sud = np.array(cavite_sud * 255)
    numerisation_cavite_nord = np.array(cavite_nord * 255)
    numerisation_cavite_centre = np.array(bin_region * 255)

    # Normalisation des caractéristiques
    est_norm = np.sum(numerisation_cavite_est) / np.sum(bin_region)
    ouest_norm = np.sum(numerisation_cavite_ouest) / np.sum(bin_region)
    sud_norm = np.sum(numerisation_cavite_sud) / np.sum(bin_region)
    nord_norm = np.sum(numerisation_cavite_nord) / np.sum(bin_region)
    centre_norm = np.sum(numerisation_cavite_centre) / np.sum(bin_region)

    return [est_norm, ouest_norm, sud_norm, nord_norm, centre_norm]

# Chargement et conversion de l'image
image_path = '/content/sample_data/ooo.jpg'  # Mettez à jour ce chemin selon votre emplacement de l'image
img = Image.open(image_path)
if img.mode != 'L':
    img = img.convert('L')

# Définition des seuils
threshold1 = 100
threshold2 = 150

# Binarisation des images
binary_image1 = binarize_image(img, threshold1)
binary_image2 = binarize_image(img, threshold2)

# Combinaison des images binarisées
final_binary_image = combine_binarizations(binary_image1, binary_image2)

# Étiquetage des régions connexes
final_binary_array = np.array(final_binary_image)
labeled_image, num_features = label(final_binary_array > 0)

# Fonction pour sélectionner une ou plusieurs régions
def bwselect(labeled_img, region_indices):

    """
    Sélectionne une ou plusieurs régions dans une image étiquetée.

    Paramètres:
    labeled_img (ndarray): L'image étiquetée.
    region_indices (list): Liste des indices des régions à sélectionner.

    Retourne:
    ndarray: Masque des régions sélectionnées.
    """

    selected_mask = np.isin(labeled_img, region_indices)
    return selected_mask

# Fonction pour déterminer les coordonnées des chiffres
def get_region_coords(labeled_img, region_label):

    """
    Obtient les coordonnées de la boîte englobante d'une région étiquetée.

    Paramètres:
    labeled_img (ndarray): L'image étiquetée.
    region_label (int): Le label de la région à extraire.

    Retourne:
    tuple: Coordonnées min_row, min_col, max_row, max_col de la boîte englobante.
    """

    region_mask = (labeled_img == region_label)
    coords = np.column_stack(np.where(region_mask))
    if coords.size == 0:
        return None, None, None, None
    min_row, min_col = coords.min(axis=0)
    max_row, max_col = coords.max(axis=0)
    return min_row, min_col, max_row, max_col

# Création du dossier pour sauvegarder les cavités
output_dir = 'data'
os.makedirs(output_dir, exist_ok=True)

# Affichage des paramètres de chaque région et des régions isolées
plt.figure(figsize=(15, 5))
for region in range(1, num_features + 1):
    bin_region = (labeled_image == region)
    min_row, min_col, max_row, max_col = get_region_coords(labeled_image, region)
    if min_row is None:
        continue
    region_image = final_binary_array[min_row:max_row+1, min_col:max_col+1]

    # Affichage de chaque région
    plt.subplot(1, num_features, region)
    plt.imshow(region_image, cmap='gray')
    plt.axis('off')
    plt.title(f'Région {region}')

    # Calcul des paramètres (ajouter d'autres paramètres si nécessaire)
    area = bin_region.sum()
    print(f"Région {region} - Surface: {area}")

plt.tight_layout()
plt.show()

# Extraction des caractéristiques normalisées pour chaque région détectée
all_features = []
for region in range(1, num_features + 1):
    print(f"Extraction des cavités pour la région {region}")
    features = extract_cavities(labeled_image, region, output_dir)
    all_features.append(features)

# Sauvegarde des caractéristiques normalisées dans un fichier
```

ETAPE D'ENTRAINEMENT DU MODELE ET PREDICTION

Tout d'abord nous avons utilisé la dataset  MNIST pour entrainer le modèle

Les deux modèles implémenter sont le CNN et le KNN. Mais nous avons utilisé le KNN pour la prediction. Après avoir fait un grisearch pour faire la recherche des paramètres optimales pour que la predition soit beauoups plus precis

```python


# Charger les données MNIST
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalisation des images
x_train = x_train / 255.0
x_test = x_test / 255.0


# Augmentation des données
datagen = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1
)
datagen.fit(x_train.reshape(-1, 28, 28, 1))
import matplotlib.pyplot as plt

def plot_training_images(x_train, y_train, num_images=20):
    plt.figure(figsize=(10, 6))
    for i in range(num_images):
        plt.subplot(4, 5, i + 1)
        plt.imshow(x_train[i], cmap='gray')
        plt.title(f'Label: {y_train[i]}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Utilisation de la fonction pour afficher des exemples d'images d'entraînement
plot_training_images(x_train, y_train, num_images=20)


# Reshape the data for CNN input
x_train_cnn = x_train.reshape(-1, 28, 28, 1)
x_test_cnn = x_test.reshape(-1, 28, 28, 1)

# Modèle CNN
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(datagen.flow(x_train_cnn, y_train, batch_size=62), epochs=10, validation_data=(x_test_cnn, y_test))

# Fonction pour extraire les caractéristiques des cavités et des moments
def extract_features(image):
    features = []

    # Binarisation de l'image
    _, binary_image = cv2.threshold(image, 0.5, 1.0, cv2.THRESH_BINARY)
    binary_image = binary_image.astype(np.uint8)

    # Propriétés des régions
    labeled_image = label(binary_image)
    regions = regionprops(labeled_image)

    if len(regions) > 0:
        region = regions[0]

        # Rapport d'aspect
        minr, minc, maxr, maxc = region.bbox
        aspect_ratio = (maxr - minr) / (maxc - minc)
        features.append(aspect_ratio)

        # Solidité
        solidity = region.solidity
        features.append(solidity)

        # Moments Hu
        moments = cv2.moments(binary_image)
        hu_moments = cv2.HuMoments(moments).flatten()
        features.extend(hu_moments)

        # Moments centraux
        central_moments = [
            moments['nu02'], moments['nu03'], moments['nu11'],
            moments['nu12'], moments['nu20'], moments['nu21'], moments['nu30']
        ]
        features.extend(central_moments)
    else:
        features.extend([0]*15)

    return np.array(features)

# Appliquer à toutes les images d'entraînement et de test
x_train_features = np.array([extract_features(img) for img in x_train])
x_test_features = np.array([extract_features(img) for img in x_test])

# Standardiser les caractéristiques
scaler = StandardScaler()
x_train_features = scaler.fit_transform(x_train_features)
x_test_features = scaler.transform(x_test_features)

# Optimisation des hyperparamètres KNN
param_grid = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=3, scoring='accuracy')
grid_search.fit(x_train_features, y_train)

best_knn = grid_search.best_estimator_
print(f"Meilleurs paramètres KNN : {grid_search.best_params_}")

# Évaluation du modèle optimisé
y_pred = best_knn.predict(x_test_features)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")

# Fonction pour prédire un chiffre dans une image
def predict_digit(image, knn_model):
    features = extract_features(image)
    features = scaler.transform([features])
    prediction = knn_model.predict(features)
    return prediction[0]

# Fonction pour extraire les chiffres de l'image
def extract_digits(image):
    _, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)
    num_labels, labeled_image = cv2.connectedComponents(binary_image)

    digits = []
    for i in range(1, num_labels):
        region = (labeled_image == i).astype(np.uint8)
        y, x = np.where(region)
        min_x, max_x = np.min(x), np.max(x)
        min_y, max_y = np.min(y), np.max(y)
        digit = binary_image[min_y:max_y+1, min_x:max_x+1]
        digit = transform.resize(digit, (28, 28), mode='constant')
        digits.append(digit)

    return digits

# Fonction pour traiter l'image et afficher les prédictions
def process_and_predict(image, knn_model, output_path='predictions.png'):
    if image is None:
        print("Erreur: Impossible de charger l'image")
        return

    digits = extract_digits(image)
    if not digits:
        print("Aucun chiffre n'a été extrait de l'image")
        return

    fig, axes = plt.subplots(1, len(digits), figsize=(10, 3))
    if len(digits) == 1:
        axes = [axes]

    for ax, digit in zip(axes, digits):
        prediction = predict_digit(digit, knn_model)
        ax.imshow(digit, cmap='gray')
        ax.set_title(f'Predicted: {prediction}')
        ax.axis('off')

    plt.tight_layout()
    plt.savefig(output_path)
    print(f'Prédictions sauvegardées à {output_path}')

# Exemple d'utilisation avec une image externe
external_image_path = '/content/sample_data/ooo.jpg'  # Remplacez par le chemin de votre image
external_image = cv2.imread(external_image_path, cv2.IMREAD_GRAYSCALE)

# Traiter et prédire les chiffres dans l'image externe
process_and_predict(external_image, best_knn)
```
